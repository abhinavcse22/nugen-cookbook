# Simple FAQ Assistant with Nugen

This notebook demonstrates how to build a basic FAQ assistant using the Nugen API.


```python
import os
import requests
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("NUGEN_API_KEY")
BASE_URL = "https://api.nugen.in"

print("API key loaded:", bool(API_KEY))
```

```python
def ask_faq(question):
    payload = {
        "model": "nugen-flash-instruct",
        "prompt": f"You are a helpful FAQ assistant.\nQuestion: {question}\nAnswer:",
        "max_tokens": 150,
        "temperature": 0.3,
        "stream": False
    }

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    response = requests.post(
        f"{BASE_URL}/api/v3/inference/completions",
        json=payload,
        headers=headers,
        timeout=60
    )

    response.raise_for_status()
    return response.json()["choices"][0]["text"]
```

```python
question = "How can I reset my account password?"
answer = ask_faq(question)

print("Question:", question)
print("\nAnswer:")
print(answer)
```
